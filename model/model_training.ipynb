```python
import pandas as pd
from sklearn import RandomForestClassifier
```

```python
# Importing data
pokemon_data = pd.read_csv("../data/pokemon.csv")
combats = pd.read_csv("../data/combats.csv")
```

```python
# Data Cleaning
pokemon_data["Type 2"] = pokemon_data["Type 2"].fillna("None")
# it's missing the 63th pokemon, and it's primeape
pokemon_data.loc[62, "Name"] = "Primeape"
```

```python
pokemon_data.head()
```

## Merging Data

```python
pokemon_data = pokemon_data.set_index("#")
df_to_merge = pokemon_data.drop(["Name", "Generation"], axis=1)
```

```python
new_df = combats.merge(
    df_to_merge, left_on="First_pokemon", right_on="#", suffixes=["_1", "_1"]).merge(
        df_to_merge, left_on="Second_pokemon", right_on="#",suffixes=["_1", "_2"]
    )
new_df
```

## Encoding Data


```python
# Encoding data
data_encoded = pd.get_dummies(new_df, columns=["Type 1_1", "Type 2_1", "Type 1_2", "Type 2_2"],drop_first=True)
```
